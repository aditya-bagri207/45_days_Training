```python
class InvalidEmailError(Exception):
    pass

try:
    email = input("Enter your email: ")

    if "@" not in email or "." not in email:
        raise InvalidEmailError("Invalid Email Address!")

    print("Valid Email Address")

except InvalidEmailError as e:
    print("Error:", e)
```
