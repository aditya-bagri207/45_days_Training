INHERITANCE 

```python
#SINGLE INHERITANCE

class company:
    pass
class Developer(company):
    pass
```

```python
class Company:
    company_name = "mahindra"
    company_add = "pilani"
class Developer(Company):
    developer_name = "aditya"
dev=Developer()      
```

```python
dev.company_name
```

```python
#multiple inheritance
class Company:
    company_name = "mahindra"
    company_add = "pilani"

class client:
    client_name = "rajesh"

class Developer(Company, client):
    developer_name = "Aditya"

dev = Developer()
print(dev.company_name)
print(dev.company_add)
print(dev.client_name)
    

```

```python
#multilevel inheritance

class person:
    name = "aditya"
    age =  "10"

class student(person):
    roll_no = "101"

class marks(student):
    marks = "80"

dev = marks() 

print(dev.name)
print(dev.age)
print(dev.roll_no)
print(dev.marks)
```

```python
#hierichical inheritance
class Parent:
    car = "BMW"

class Son(Parent):
    pass

class Daughter(Parent):
    pass

print(Son.car)
print(Daughter.car)
```

```python
#hybrid inheritance

class Grandparent:
    land = 50

class Father(Grandparent):
    pass

class Mother(Grandparent):
    pass

class Child(Father, Mother):
    pass

c = Child()

print(c.land)
```

```python
# encapsulation in python

class Company:
    company_name = "google" #public variable
    _company_add = "jaipur"  #protected
    __company_acc = "7976314858" #private

    def show(self):
        print(self.__company_acc)

```
