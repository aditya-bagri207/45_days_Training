NumPy (Numerical Python) is a powerful Python library used for numerical computing. It provides support for:

Multi-dimensional arrays (ndarray)
Mathematical operations on arrays
Linear algebra
Statistical functions
Random number generation
Fast execution compared to Python lists

```python
import numpy as np

arr = np.array([1,2,3])
print(arr)
```

```python
print(np.__version__)
```

```python
#2d arrays

arr2 = np.array([[1,2],[3,4]])
print(arr2)
```

```python
lst = [1,2,3,4,5]
print("This is my lst",lst)
print("Type of my list: ", type(lst))
print("list is converted into array")
arr = np.array(lst)
print("this is array",arr)
print("type of my array", type(arr))
```

```python
arr = np.zeros(2)
print(arr)
```

```python
arr = np.ones(3)
print(arr)
```

```python
arr = np.zeros([3,3])
arr
```

```python
arr = np.arange(60)
print(arr)
arr = arr.reshape(12,5)
print(arr)
```

```python
arr = np.array([[1,3],[4,5]])
print(arr)
print(arr.shape)   
print(arr.ndim)    
print(arr.size)    
print(arr.dtype)   
```

```python
#sliceing

arr = np.array([1,2,3,4,5])
print(arr[0:2])
```

```python
arr[::-1]
arr
```

```python
arr = np.array([10,20,30])
print(arr[arr>20])
```
