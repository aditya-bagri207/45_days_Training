Machine Learning (ML) is a branch of Artificial Intelligence (AI) that enables computers to learn patterns from data and make predictions or decisions without being explicitly programmed.

Types of Machine Learning

Supervised Learning – Trains on labeled data (e.g., predicting house prices).
Unsupervised Learning – Finds patterns in unlabeled data (e.g., customer segmentation).
Reinforcement Learning – Learns through rewards and penalties (e.g., game-playing AI).

Key Terms

Dataset: Collection of data used for training.
Features: Input variables.
Label/Target: Output variable.
Training: Teaching the model using data.
Testing: Evaluating model performance on unseen data.
Model: Algorithm that learns from data.

Common Algorithms

Linear Regression
Logistic Regression
Decision Tree
Random Forest
K-Nearest Neighbors (KNN)
Support Vector Machine (SVM)


Applications

Recommendation systems
Spam detection
Image recognition
Medical diagnosis
Fraud detection
