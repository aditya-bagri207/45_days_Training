Unsupervised Learning is a type of Machine Learning in which the model learns from unlabeled data. It finds patterns, 
relationships, or groups in the data without being told the correct answers.

types of unsupervised learning

1. Clustering
Groups similar data points together.
No labels are provided.
Examples of algorithms:
K-Means Clustering
Hierarchical Clustering
DBSCAN

Example: Grouping customers based on their shopping habits.

2. Association
Finds relationships or rules between items in a dataset.
Examples of algorithms:
Apriori
FP-Growth

Example: If a customer buys bread, they are also likely to buy butter (market basket analysis).

3. Dimensionality Reduction
Reduces the number of features while keeping the important information.
Makes data easier to analyze and visualize.
Examples of algorithms:
Principal Component Analysis (PCA)
t-SNE

Example: Compressing a dataset with 100 features into 10 important features.
